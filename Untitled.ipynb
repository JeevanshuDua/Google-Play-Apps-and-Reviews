# Sentiment analysis of the reviews

```python
df = pd.read_csv('googleplaystore_user_reviews.csv')
```

```python
df
```

```python
plt.subplots(figsize=(18,8))
sns.heatmap(df.isnull())
```

```python
df = df.dropna()
#df = df.drop(['App'], axis = 1)
```

```python
df
```

```python
sns.countplot(df.Sentiment)
```

```python
(df[['App', 'Translated_Review']].groupby('App').count().sort_values(by='Translated_Review'))
```

```python
pd.crosstab(df.App, df.Sentiment)
```

```python
df_negative = df[df['Sentiment']=='Negative']
words = ' '.join(df_negative['Translated_Review'])
cleaned_word = " ".join([word for word in words.split()
                            
                            ])
wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(cleaned_word)


plt.figure(1,figsize=(12, 12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
